```python
from abc import ABC, abstractmethod

class Library(ABC):
    @abstractmethod
    def borrow_book(self, user):
        pass

    @abstractmethod
    def return_book(self, user):
        pass

# Book class encapsulating book details and borrowing logic
class Book:
    def __init__(self, title, author, ISBN):
        self.__title = title
        self.__author = author
        self.__ISBN = ISBN
        self.__is_available = True  # Default value is available
    
   
    @property
    def title(self):
        return self.__title


    @property
    def author(self):
        return self.__author


    @property
    def ISBN(self):
        return self.__ISBN

    
    @property
    def is_available(self):
        return self.__is_available

    # Method to borrow the book
    def borrow(self):
        """
        Marks the book as borrowed (sets is_available to False) if available.
        
        Returns:
        - True if the book was successfully borrowed, False if not available.
        """
        if self.__is_available:
            self.__is_available = False
            return True
        return False

    # Method to return the book
    def return_book(self):
        """
        Marks the book as available (sets is_available to True).
        """
        self.__is_available = True

# Member class inheriting from Library
class Member(Library):
    def __init__(self, name, member_id):
        self.__name = name
        self.__member_id = member_id
        self.__borrowed_books = []  # Initializes an empty list for borrowed books

    @property
    def name(self):
        return self.__name

    @property
    def member_id(self):
        return self.__member_id

    @property
    def borrowed_books(self):
        return self.__borrowed_books

    # Method to borrow a book
    def borrow_book(self, book):
        """
        Allows the member to borrow a book if it is available.

        Parameters:
        - book (Book): The book to be borrowed.
        """
        if book.borrow():
            self.__borrowed_books.append(book)
            print(f"{self.__name} has borrowed '{book.title}'")
        else:
            print(f"'{book.title}' is not available for borrowing")

    # Method to return a borrowed book
    def return_book(self, book):
        if book in self.__borrowed_books:
            book.return_book()
            self.__borrowed_books.remove(book)
            print(f"{self.__name} has returned '{book.title}'")
        else:
            print(f"{self.__name} doesn't have '{book.title}' borrowed")

    # Method to get member details
    def get_details(self):
        return f"Member: {self.__name}, ID: {self.__member_id}"

# Staff class inheriting from Member
class Staff(Member):
    def __init__(self, name, member_id, position):
        super().__init__(name, member_id)
        self.__position = position

    # Override get_details to include position
    def get_details(self):
        return f"Staff: {self.name}, ID: {self.member_id}, Position: {self.__position}"

# Librarian class inheriting from Staff
class Librarian(Staff):
    def __init__(self, name, member_id, position):
        super().__init__(name, member_id, position)
        self.__library_books = []

    # Method for librarian to add books to the library
    def add_book(self, book):
        self.__library_books.append(book)
        print(f"Book '{book.title}' by {book.author} added to the library")
```

```python
book1 = Book("1984", "George Orwell", "123456789")
book2 = Book("To Kill a Mockingbird", "Harper Lee", "987654321")
member = Member("John Doe", "M001")
staff = Staff("Alice Smith", "S001", "Librarian")
librarian = Librarian("Bob Brown", "L001", "Head Librarian")
```
